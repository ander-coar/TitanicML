```python
import  pandas as pd
from IPython.testing.tools import printed_msg

serie_edad = pd.Series([10, 12, 14, 15, 17.5, 20, 22, 25, 30, 80]) # el dato faltante era 18 años
serie_salario = pd.Series([10.000, 12.000, 14.0000, 15.0000, 18.0000, 20.0000, 22.000, 25.000, 30.000, 80.000])
serie_edad.head(10)
```

```python
serie_salario.head(10)
```

```python
print(f"el promedio de edades es: {serie_edad.mean()}")
print(f"el promedio de salarios es: {serie_salario.mean()}")
print(f"la mediana  de edades es: {serie_edad.median()}")
print(f"el la mediana de salarios es: {serie_salario.median()}")
```

```python
# Limpiar datos
# usamos el metodo IQR para detectar outliers en la serie de edades y en el salario
# interpolación lineal
q1 = serie_edad.quantile(0.25)
q3 = serie_edad.quantile(0.75)
iqr = q3 - q1

print(f"q1 = {q1} años -  q3= {q3} años - IQR= {iqr} amplitud de años")

# El iqr no es una medida es una AMPLITUD
# el 50% de los datos se encuentran entre q1 y q3
# y hay una dispersión de 11 años entre q1 y q3
# El propósito del IQR es detectar outliers

margen = 1.5 * iqr
li = q1 - margen
ls = q3 + margen

print(f"Limite inferior: {li} años")
print(f"Limite superior: {ls} años")
```

```python
datos_limpios_edad = serie_edad[(serie_edad >= li) & (serie_edad <= ls)]
datos_limpios_edad.head(10)
```

```python
print(f"el promedio con los datos limpios es: {datos_limpios_edad.mean()}")
print(f"la mediana con los datos limpios es: {datos_limpios_edad.median()}")
```

```python
# STD - Desviación estándar
# USamos el STD de pandas que calcula la desviación estándar de una serie o dataset

std_datos_sucios = serie_edad.std()
print(f"La desviación estándar es: {std} años")

std_datos_limpios = datos_limpios_edad.std()
print(f"La desviación estándar con los datos limpios es: {std: .2f} años")
# el promedio es de 18.5 años en los datos limpios
# Las edades varían en un rango de +- std 7 años respecto al promedio
```

```python
# 3 Coeficiente de variación (CV)
# El coeficiente de variación es una medida de dispersión relativa que se expresa como un porcentaje.
# CV = (STD / Media) * 100

cv_datos_sucios = (std_datos_sucios / serie_edad.mean()) * 100
cv_datos_limpios = (std_datos_limpios/ datos_limpios_edad.mean()) * 100

print(f"El coeficiente de variación con datos sucios es: {cv_datos_sucios: .2f} %")
print(f"El coeficiente de variación con datos limpios es: {cv_datos_limpios: .2f} %")

# Tabla de decsisión
# CV < 15%: Baja dispersión - Usar la media
# Entre 15% - 30%: Moderada dispersión - Usar la mediana
# CV > 30%: Alta dispersión - imputación por grupos, si no tenemos grupos de variables usar la mediana

# El CV nos ayuda a generar ML robustos y confiables
```
