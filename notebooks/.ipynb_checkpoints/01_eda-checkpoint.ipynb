```python
# En python no usamos la mediana para hallar outliers, usamos el rango intercuartÃ­lico (IQR)
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
print("q1 es =", q1)
print("q3 es = ", q3)
iqr = q3 - q1
print("EL IQR es:", iqr)
```

```python
margen = 1.5 * iqr
li = q1 - margen
ls = q3 + margen
print("Margen: ", margen)
print("Limite inferior", li)
print("Limite superior", ls)
```

```python
print(f"LA edad minima es: {df['Age'].min()}")
print(f"LA edad maxima es: {df['Age'].max()}")
```

```python
outliers_edad = df[(df['Age'] < li ) | (df['Age'] > ls)] # operaciÃ³n boleana para filtrar los outliers
# En pandas el simbolo or es | y and es &
# una serie de pandas es una columna del DataFrame
print(outliers_edad['Age'].tolist())  # Muestra las filas donde 'Age' es un outlier

# SE debe verificar si hay outliers en las columnas numericas del dataset
# Porque para llenar los valores faltantes si hay outliers no se debe usar la media (Promedio), sino la mediana
```

```python
import matplotlib.pyplot as plt
# Diagrama de caja (boxplot) para visualizar los outliers
fig, ax = plt.subplots(figsize=(16, 5))
bp = ax.boxplot(df['Age'].dropna(),
                vert = False, # Coloca el eje horizontal
                patch_artist = True, # Rellena el boxplot con color, Rojo por defecto
                widths = 0.8, # Ancho del de la caja del eje vertical
                # Darle propiedades a la caja- facecolor es el color de relleno, alpha es la transparencia de ese relleno
                boxprops = dict(facecolor = 'lightblue', alpha = 0.5 ), # Propiedades de la caja
                medianprops = dict(color = 'red', linewidth = 1, linestyle="--"), # Propiedades de la linea de la mediana
                whiskerprops = dict(color = 'red', linewidth = 2, linestyle="-"), # Propiedades de los bigotes
                #Caps es tapas de los bigotes
                capprops = dict(color = 'blue', linewidth = 5, linestyle="-"), # Propiedades de las tapas de los bigotes
                #flier es lo que vuela fuera de los bigotes (outliers)
                # marker puede recibir diferentes simbolos como 'o', 's', '^', 'x', 'd', '*', 'D', 'X', '+', etc
                flierprops=dict(marker='s', markersize=8, markerfacecolor='black', markeredgecolor='white', alpha = 0.5 ),
                )

ax.axvline(q1, color='gray', linestyle='--', linewidth=2, label=f'Q1 = {q1:.0f}')
ax.axvline(q3, color='green', linestyle='--', linewidth=2, label=f'Q3 = {q3:.0f}')
ax.axvline(li, color='orange', linestyle=':', linewidth=2, label=f'Limite Inferior = {li:.2f}')
ax.axvline(ls, color='orange', linestyle=':', linewidth=2, label=f'Limite Superior = {ls:.2f}')
ax.axvline(iqr, color='red', linestyle=':', linewidth=5, label=f'IQR = {iqr:.2f}')
ax.legend(loc='best', fontsize='12')
ax.set_title('Edad con Outliers de los pasajeros del Titanic', fontsize=16)
ax.set_xlabel('Edad', fontsize=14)
ax.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.show()


```

```python
# conteo de personas por grupos de edad
# para hacerlo puedo usar pd.cut para crear categorias de edad y bins para seccionar o definir los rangos
# o se puede usar filtrado booleano y contar con value_counts
age_bins = [0, 20, 38, 100]
age_labels = ['Grupo 1: de 0 a 20 aÃ±os', 'Grupo 2: de 20 a 38 aÃ±os', 'Grupo 3: de 38 a 100 aÃ±os']
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False    ) # el Right=False indica que el limite derecho no se incluye, el limite derecho hace referencia a el ultimo valor del bin
age_group_counts = df['AgeGroup'].value_counts().sort_index()
print(age_group_counts)

# a diferencia de sum value_counts no ignora los NaN
```

```python
# usando el filtrado booleano
grupo_1 = (df['Age'] < 20).sum()
grupo_2 = ((df['Age'] >= 20) & (df['Age'] < 38)).sum()
grupo_3 = (df['Age'] >= 38).sum()
print(f"Grupo 1: de 0 a 20 aÃ±os: {grupo_1}")
print(f"Grupo 2: de 20 a 38 aÃ±os: {grupo_2}")
print(f"Grupo 3: de 38 a 100 aÃ±os: {grupo_3}")
```
