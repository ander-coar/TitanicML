```python
# Outliers en la columna Age
# En python no usamos la mediana para hallar outliers, usamos el rango intercuart铆lico (IQR)
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
print("q1 es =", q1)
print("q3 es = ", q3)
iqr = q3 - q1 # el iqr es la amplitud entre el tercer y primer cuartil
print("EL IQR es:", iqr)
```

```python
margen = 1.5 * iqr
li = q1 - margen
ls = q3 + margen
print("Margen: ", margen)
print("Limite inferior", li)
print("Limite superior", ls)
```

```python
print(f"LA edad minima es: {df['Age'].min()}")
print(f"LA edad maxima es: {df['Age'].max()}")
```

```python
outliers_edad = df[(df['Age'] < li ) | (df['Age'] > ls)] # operaci贸n boleana para filtrar los outliers
# En pandas el simbolo or es | y and es &
# una serie de pandas es una columna del DataFrame
print(outliers_edad['Age'].tolist())  # Muestra las filas donde 'Age' es un outlier

# SE debe verificar si hay outliers en las columnas numericas del dataset
# Porque para llenar los valores faltantes si hay outliers no se debe usar la media (Promedio), sino la mediana
```

```python
# Sexo, Clase, Titulo, Tarifa, Embarque
# Usaremos el metodo isin para filtrar los datos
lista_titulos_nulos = ['Mr', 'Mrs', 'Miss', 'Master', 'Dr']
df_medianas = df[df['Title'].isin(lista_titulos_nulos)].groupby(['Title','Pclass','Embarked'])['Age'].median().reset_index()
df_medianas
```

```python
#merge se encarga de unir dos dataframes por una columna en com煤n
df_nulos =  df[(df['Title'].isin(lista_titulos_nulos)) & (df['Age'].isna())].groupby(['Title','Pclass','Embarked']).size().reset_index(name='Cantidad')
df_nulos
```

```python
# Comparar los dataframes df_nulos y df_medianas
# En el tipo de merge existen 3 tipos de comparaci贸n es both, letf_only, right_only
comparacion = df_nulos.merge(df_medianas, on=['Title', 'Pclass','Embarked'], how='left', indicator = True)
comparacion
```

```python
# axis es la direci贸n en la que se puede aplicar una funci贸n
# axis = 1 o axis = 0
# axis = 1 aplica la funci贸n a lo largo de las columnas
# axis = 0 aplica la funci贸n a lo largo de las filas
```

```python
#apply recorre de forma autonoma cada fila o columna no es necesario crear un for

def imputar_edad(row):
    # si la fila tiene edad pues vamos a devolver la misma edad, si no tiene edad vamos a buscar la edad mediana correspondiente
    if pd.notna(row['Age']):
        return row['Age']

    # Si no tiene edad, buscamos la mediana correspondiente
    filtro = ((df_medianas['Title'] == row['Title']) & (df_medianas['Pclass'] == row['Pclass']) & (df_medianas['Embarked'] == row['Embarked']))

    mediana = df_medianas.loc[filtro, 'Age'].values[0] # obtenemos el valor de la mediana
    return mediana

# aplicar la funci贸n a cada fila del DataFrame
df['Age'] = df.apply(imputar_edad, axis=1)

df['Age'].isna().sum()
```
