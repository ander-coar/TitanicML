```python
# En python no usamos la mediana para hallar outliers, usamos el rango intercuartÃ­lico (IQR)
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
print("q1 es =", q1)
print("q3 es = ", q3)
iqr = q3 - q1 # el iqr es la amplitud entre el tercer y primer cuartil
print("EL IQR es:", iqr)
```

```python
margen = 1.5 * iqr
li = q1 - margen
ls = q3 + margen
print("Margen: ", margen)
print("Limite inferior", li)
print("Limite superior", ls)
```

```python
print(f"LA edad minima es: {df['Age'].min()}")
print(f"LA edad maxima es: {df['Age'].max()}")
```

```python
outliers_edad = df[(df['Age'] < li ) | (df['Age'] > ls)] # operaciÃ³n boleana para filtrar los outliers
# En pandas el simbolo or es | y and es &
# una serie de pandas es una columna del DataFrame
print(outliers_edad['Age'].tolist())  # Muestra las filas donde 'Age' es un outlier

# SE debe verificar si hay outliers en las columnas numericas del dataset
# Porque para llenar los valores faltantes si hay outliers no se debe usar la media (Promedio), sino la mediana
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
# from matplotlib import pyplot as plt

# esto es una tupla
# fig es lienzo que va a contener los graficos
# ax es el eje donde se van a dibujar los graficos
fig, ax = plt.subplots(figsize=(8,5))  # crea una figura y un eje para graficar de 8x6 pulgadas

bp = ax.boxplot(df['Age'].dropna(), vert = False, patch_artist = True, widths = 0.8,
                boxprops = dict(facecolor = 'red', alpha = 0.8),
                medianprops = dict(color = 'black', linewidth = 2, linestyle = ':'),
                whiskerprops = dict(color = "red", linewidth = 3, linestyle = '-'),
                capprops = dict(color = 'red', linewidth = 1.5, linestyle = '-'),
                flierprops = dict(markerfacecolor = 'blue', markeredgecolor = 'black', marker = 'D', markersize = 12, alpha = 0.5)
                )
ax.axvline(q1, color = 'green', linestyle = '--', linewidth = 1.5, label = f'Q1: {q1:.0f} ')# agrega lineas verticales para el tercer cuartil
ax.axvline(q3, color = 'orange', linestyle = '--', linewidth = 1.5, label = f'Q3: {q3:.0f} ') # agrega lineas verticales para el tercer cuartil
ax.axvline(li, color = 'blue', linestyle = '--', linewidth = 1.5, label = f'li: {li:.2f} ') # agrega lineas verticales para el limite inferior
ax.axvline(ls, color = 'blue', linestyle = '--', linewidth = 1.5, label = f'ls: {ls:.2f} ') # agrega lineas verticales para el limite superior
ax.legend(loc = 'best', fontsize = 10) # agrega una leyenda al grafico
ax.set_title("Diagrama de caja de la edad de los pasajeros del Titanic", fontsize = 14) # titulo del grafico
ax.set_xlabel("Edad de los pasajeros del Titanic", fontsize = 12) # etiqueta del eje x
ax.grid(True, alpha = 0.2, axis='x') # activa la cuadricula en el grafico
ax.grid(True, alpha = 0.2, axis='y') # activa la cuadricula en el grafico

ax.xaxis.set_major_locator(MultipleLocator(5)) # Establece los marcadores principales del eje x cada 5 unidades
plt.tight_layout() # ajusta el diseÃ±o para que no se corten los elementos
plt.show() # es para mostrar el grafico en pantalla



```
